```python
import pandas as pd
import sklearn
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("jpm444_hw_data2.csv")
```

Calculate the Macro Average

```python
sklearn.metrics.precision_score(df[['label_true']], df[['label_assigned']],average='macro')
```

```python
cm=confusion_matrix(df[['label_true']],df[['label_assigned']],labels=['economy','human_interest','politics']) # True label is in the first position
cm
```

```python
matrix = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['economy','human_interest','politics'])
matrix.plot()
```

Export the figure as png

```python
matrix.plot().figure_.savefig('confusion_matrix_hw2.png')
```
