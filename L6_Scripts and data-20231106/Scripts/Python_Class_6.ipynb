##Looping over many search terms

Load the dictionary

```python
mydict = pd.read_csv("/content/dictionary.txt", sep="\t")
mydict
```

Load the texts to analyze

```python
#df = pd.read_csv("/content/ukraine_kwics_2018_2023_sample1000.txt", sep="\t")
df = pd.read_csv("/content/drive/MyDrive/vyuka_a_vedeni_praci/NIA_ML/Class6/Data/ukraine_kwics_2018_2023_sample1000.txt", sep="\t")
df.head()
```

Create a list of regex patterns, lowercased, with word boundaries

```python

list_of_regex=mydict['pattern'].tolist()
list_of_regex=[x.lower() for x in list_of_regex]
list_of_regex=["".join(['\\b',x,'\\b']) for x in list_of_regex] #add word boundaries, not the double \\
list_of_regex
```

Create a corresponding list of entity labels to use as variable names

```python
list_of_regex_labs =mydict['entity3n'].tolist()
list_of_regex_labs
```

Loop over individual rows (main loop, iterrows) and over the search terms

```python
for index, row in df.iterrows():
    text=row['text']

    #iterate over patterns you look for
    i=0
    while i<len(list_of_regex_labs):
      regexlab=list_of_regex_labs[i]
      regex=list_of_regex[i]
      i=i+1

      res=len(re.findall(regex, text))

      #store the result to the given row and column with the label as the name
      df.loc[index, regexlab] = res
```
