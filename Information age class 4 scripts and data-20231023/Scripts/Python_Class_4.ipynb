## Parsing an RSS feed

```python
!pip install feedparser
```

```python
import feedparser
import json
import requests
import pandas as pd
```

Get the link to the feed you are interested in and extract useful elements

```python

d = feedparser.parse('http://rss.cnn.com/rss/edition.rss')
#d['feed']['title']
d['feed']['date']
```

```python
d
```

```python
d.entries[0]
```

Create a data frame with all entries

```python
entrieslist = []
for entry in d.entries:
    entrieslist.append((entry.title, entry.link,d['feed']['date']))
df = pd.DataFrame(entrieslist, columns=['article_title', 'url','date'])
```

## BeautifulSoup to extract content from html

```python
#!pip install bs4
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.bbc.com/news/world-middle-east-67188346')
soup = BeautifulSoup(r.text, "html.parser")
```

```python
text=soup.get_text()
```

```python
print(text)
```

```python
outputlist = []
for entry in soup.find_all('p'):
  #print(entry)
  #print(entry.contents)
  outputlist.append(entry.contents)
```

```python
outputlist
```

```python
outputlist2=flat_list = [item for sublist in outputlist for item in sublist] #some list elements had two elements, so flatten the list
```

```python
df = pd.DataFrame(outputlist2, columns=['paragraph_text']) #convert to data frame
```
